```python
with open(f'{out_dir}{out_name}.qmd','r', encoding="utf-8") as S:
    Slides = S.read()

for img in os.listdir(out_dir+'images/'):
    for img2 in os.listdir(out_dir+'images/'):
        if img2 != img:
            if open(out_dir+'images/'+img,"rb").read() == open(out_dir+'images/'+img2,"rb").read():
                print('Duplicate:', img2)
                Slides=Slides.replace(img2,img)


for img in os.listdir(out_dir+'images/'):
    if img not in Slides:
        print(f'Deleting {img}')
        os.remove(out_dir+'images/'+img)


for dat in os.listdir(out_dir+'Data/'):
    for dat2 in os.listdir(out_dir+'Data/'):
        if dat2 != dat:
            if open(out_dir+'Data/'+dat,"rb").read() == open(out_dir+'Data/'+dat2,"rb").read():
                print('Duplicate:', dat,dat2)
                Slides=Slides.replace(dat2,dat)

for dat in os.listdir(out_dir+'Data/'):
    if dat not in Slides:
        print(f'Deleting {dat}')
        os.remove(out_dir+'Data/'+dat)

with open(f'{out_dir}{out_name}.qmd','w', encoding="utf-8") as out:
    out.write(Slides)
```
